### **¿Qué es un púlsar?**

Los pulsares son **estrellas de neutrones** altamente magnetizadas, que giran rápidamente sobre su propio eje. Aquellos que son jóvenes, rotan considerablemente más rápido que los longevos, puesto que pierden energía con el paso del tiempo. Esta pérdida de energía es objeto de estudio y es lo que analizaremos en el presente proyecto.

<div style="background-color: #f1f79d; color: black; padding: 10px; border-radius: 5px;">

### **Púlsar del Cangrejo**

El **Púlsar del Cangrejo (PSR B0531+21)** es uno de los pulsares mejor caracterizados observacionalmente. Está situado en la Nebulosa del Cangrejo, posee 25 km de diámetro y realiza 30 vueltas por segundo, o sea, su frecuencia es de 30 Hz. Su periodo de rotación disminuye 38 nanosegundos por día debido a la energía que libera. 

**¿Por qué ocurre esto?**

Si bien la rotación de este púlsar es muy estable debido a que no se aplica ninguna fuerza y su geometría no cambia, este frena con el paso del tiempo debido a la pérdida de energía. El principal mecanismo de pérdida de energía que opera sobre los pulsares es el **freno magnético**.


<div style="background-color: #f1f79d; color: black; padding: 10px; border-radius: 5px;">

### **Frenado magnetodipolar**

El modelo clásico de un púlsar describe a la estrella de neutrones como un dipolo magnético cuyo eje no coincide con el eje de rotación. Como consecuencia de esta desalineación, el momento dipolar varía en el tiempo, lo que conduce a la emisión de **radiación electromagnética**. La **potencia radiada**, equivalente a la pérdida de energía rotacional por unidad de tiempo, puede derivarse a partir de la electrodinámica clásica y depende del período de rotación y de la tasa de frenado del púlsar, ambas cantidades accesibles observacionalmente.

<div style="background-color: #d4eeff; color: black; padding: 10px; border-radius: 5px;">

### **El modelo físico**

La energía rotacional del púlsar es: 

$$

E_{rot} = \frac{1}{2}I(\Omega)(\Omega)^{2}

$$

donde 
- $I$ es el momento de inercia.
- $\Omega$ es la frecuencia angular de rotación.

Su evolución temporal se obtiene desde el balance energético: 

$$

\frac{dE_{rot}}{dt} = -P_{loss}

$$

donde $P_{loss}$ es la potencia total perdida por los procesos electrodinámicos.

### **El proyecto**

El momento de inercia se expresa en función de la frecuencia angular: 

$$

I(\Omega) = I_0(1 + \epsilon\Omega^2)

$$

La evolución del ángulo magnético es: 

$$

\alpha(t) = \alpha_0 + \beta t

$$

La potencia radiada por un dipolo magnético en rotación es: 

$$

P_{dip} = - \frac{2}{3c^{3}} (BR^{3})^{2}\Omega^{4}\sin^{2}{\alpha (t)}

$$

Por lo tanto, el **balance energético** es: 

$$

\frac{d}{dt} \left[ \frac{1}{2} I(\Omega) \Omega^2 \right] = -\frac{2B^2R^6}{3c^3} \Omega^4 \sin^2 \alpha(t)

$$

El **objetivo** del proyecto es resolver estas ecuaciones mediante los métodos numéricos RK4 y RK45, para posteriormente comparar los resultados arrojados con los valores observacionales, tomando en cuenta la estabilidad, precisión y el valor numérico del índice de frenado.

### **Preparar la ecuación diferencial**

Para poder resolver este problema numéricamente, primero debemos dejar la ecuación diferencial expresada de la siguiente forma:
$$
\frac{d \Omega(t)}{dt} = f(t,y)
$$
Sabemos que
$$
P_{dip} =  \frac{2}{3c^{3}} (BR^{3})^{2}\Omega^{4}\sin^{2}{\alpha (t)}
$$
Reemplazando en la ecuación
$$
\frac{d}{dt}\left[ \frac{1}{2}I(\Omega)\Omega^{2} \right] = -P_{dip}(t)
$$
Por definición $\large I(\Omega) = I_{0}(1+\epsilon\Omega^{2})$, reemplazando:
$$
\frac{d}{dt}\left[ \frac{1}{2}I_{0}(1+\epsilon\Omega^{2})\Omega^{2} \right] = -P_{dip}(t)
$$
Sacamos las constantes y distribuimos $\large \Omega^{2}$
$$
\begin{align}
\frac{d}{dt}[(1+\epsilon\Omega^{2})\Omega^{2}] &= -\frac{2P_{dip}}{I_{0}} \\
\frac{d}{dt}[(\Omega^{2}+\epsilon\Omega^{4})] &= -\frac{2P_{dip}}{I_{0}}
\end{align}
$$
Aplicamos la derivada usando la regla de la cadena
$$
\begin{align}
2\Omega\dot{\Omega} + 4\epsilon\Omega^{3}\dot{\Omega} &= -\frac{2P_{dip}}{I_{0}} \\
\dot{\Omega}(\Omega + 2\epsilon\Omega^{3}) &= -\frac{P_{dip}}{I_{0}}
\end{align}
$$
Finalmente obtenemos que:
$$
\boxed{\dot{\Omega} = \frac{d\Omega(t)}{dt} = -\frac{P_{dip}(t)}{I_{0}\Omega(1 + 2\epsilon\Omega^{2})}}
$$

Para poder empezar a resolver nos estarían faltando 3 datos, $\epsilon$, $\alpha_{0}$ y $\beta$.  Para encontrar estas constantes nos ayudaremos de el balance energético.

Sabemos que:
$$
\frac{\dot{\Omega}}{I_{0}\Omega(1+\epsilon 2\Omega^{2})} = -P_{dip}
$$
y también:
$$
P_{dip} = \frac{2}{3c^{3}} (BR^{3})^{2}\Omega^{4}\sin^{2}{\alpha (t)} = K\Omega^{4}\sin^{2}\alpha(t)
$$
reemplazando:
$$
{\dot{\Omega}}{I_{0}\Omega(1+\epsilon 2\Omega^{2})} = -K\Omega^{4}\sin^{2}\alpha(t)
$$
$$
\implies \sin^{2}\alpha(t) = \frac{{\dot{\Omega}}{I_{0}(1+\epsilon 2\Omega^{2})}}{K\Omega^{3}}
$$
La desigualdad $0 \le \sin^2\alpha(t) \le 1$ se cumple porque el seno de cualquier ángulo real siempre toma valores entre $-1$ y $1$. Al elevar ese valor al cuadrado, los números negativos se vuelven positivos y el máximo posible sigue siendo $1$, por lo que el resultado queda necesariamente entre $0$ y $1$. Por eso, $\sin^2\alpha(t)$ nunca puede ser negativo ni mayor que uno.

Esto nos sirve como condición para determinar constantes, si empezamos en $t =0 \implies \alpha(t=0) = \alpha_{0}$ por lo tanto:
$$
\implies \sin^{2}\alpha_{0} = \frac{{\dot{\Omega_{0}}}{I_{0}(1+\epsilon 2\Omega_{0}^{2})}}{K\Omega_{0}^{3}}
$$
y este cómo debe estar entre cero y uno sólo nos sirven valores específicos de $\epsilon$ en este caso ocuparemos $\epsilon \approx 10^{-6}$. Si ocupamos los valores que nos dieron como iniciales nos da que:
$$
\alpha_{0} = \arcsin(\sqrt{ \sin^{2} \alpha_{0}}) \approx 0,64 \ \  [rad] 
$$
Finalmente el único valor que tabularemos es $\beta$ el cual para los pulsares es típico que estén en el orden de $10^{-13}$ y $10^{-12}$.

```python
# --- Constantes Físicas (CGS) ---
I0 = 1e45              # g*cm^2
epsilon = 1e-6         # adimensional (una deformación razonable)
c = 2.9979e10          # cm/s
R = 1.2e6              # cm (12 km)
B = 3.8e12             # Gauss

# --- Parámetros de la Evolución del Ángulo ---
alpha0 = 0.64          # rad
betta = 1e-13          # rad/s

# --- Parámetros de la simulación
omega0 = 187.5         # rad/s
omega_dot0 = -2.35e-9  # rad/s^-2

print("sin^2(alpha) =", np.sin(alpha0)**2)
print("B^2 =", B**2)
print("R^6 =", R**6)

k = (2 * (B * R**3)**2) / (3 * c**3)
sin2a = -omega_dot0 * I0 * (1 + (2 * epsilon * omega0**2)) / (k * omega0**3)

print(sin2a)

a = np.arcsin(np.sqrt(sin2a))
print(a)


# --- 2. Parámetros para la Fórmula Analítica (Imagen) ---
n = 2.51               # Índice observado
t = 1e10            # 1000 años en segundos
potencia = 1 / (1 - n)

K_analitica = -omega_dot0 / (omega0**n)


# Usamos la fórmula: Omega(t) = [Omega0^(1-n) - (1-n)*K*t]^(1/(1-n))
# Para t = 0
omega_analitica0 = (omega0**(1 - n) - (1 - n) * K_analitica * 0)**potencia
# Para t = 1000 años
omega_analitica = (omega0**(1 - n) - (1 - n) * K_analitica * t)**potencia

print(f"Resultados de la Aproximación Analítica (n={n}):")
print("-" * 45)
print(f"Omega inicial (t=0):    {omega_analitica0:.4f} rad/s")
print(f"Omega final (1000 años): {omega_analitica:.4f} rad/s")
print(f"K analítica calculada:   {K_analitica:.2e}")

# ----- Calculo energía analítica
```

3. Definimos la función de la derivada respecto al tiempo de omega, para esto contruimos una función con todas las constantes necesarias.

```python
def omega_dot(t, y, B, R, c, I0, epsilon, alpha0, betta):
    # 'y' es una lista/array, extraemos el primer valor (omega)
    omega = y

    #Ángulo que depende del tiempo
    a = alpha0 + betta * t
    k = (2 * (B * R**3)**2) / (3 * c**3)

    #P_dip
    p_dip = k * (omega**4) * (np.sin(a)**2)
    denominador = I0 * omega * (1 + 2 * epsilon * (omega**2))
    
    f = -p_dip / denominador

    return f
```

4. Soluciones a la EDO

Primero empezaremos usando nuestra implementación de rk4, el cual la llamaremos de nuestro modulo

```python
import module as mod

segundos_en_un_año = 365.25 * 24 * 3600
años_simulacion = [100, 1000, 10000]

resultados_rk4 = {}

for años in años_simulacion:

    t_final = años * segundos_en_un_año
    h_estable = 24 * 3600

    solucion_t, omega_rk4 = mod.rk4_method(
        lambda t, y: omega_dot(t, y, B, R, c, I0, epsilon, alpha0, betta),
        t0=0.0,
        y0=omega0,
        h=h_estable,
        tf=t_final
    )

    resultados_rk4[años] = (solucion_t, omega_rk4)

    t_años = solucion_t / segundos_en_un_año

    plt.figure(figsize=(10,6))
    plt.plot(t_años, omega_rk4, label="RK4", color="royalblue")
    plt.title(f"RK4 - Evolución en {años} años")
    plt.xlabel("Tiempo (años)")
    plt.ylabel(r"$\Omega$ (rad/s)")
    plt.grid()
    plt.legend()
    plt.show()
```


4.2 Primero podemos intentr hacer la solución con respecto a rk45 de scipy

```python
resultados_rk45 = {}

for años in años_simulacion:

    solucion_t, _ = resultados_rk4[años]
    t_final = años * segundos_en_un_año

    omega_rk45 = solve_ivp(
        fun=omega_dot,
        t_span=(0, t_final),
        y0=[omega0],
        method="RK45",
        args=(B, R, c, I0, epsilon, alpha0, betta),
        rtol=1e-9,
        atol=1e-12,
        t_eval=solucion_t
    )

    resultados_rk45[años] = omega_rk45

    t_años = omega_rk45.t / segundos_en_un_año

    plt.figure(figsize=(10,6))
    plt.plot(t_años, omega_rk45.y[0], label="RK45", color="firebrick")
    plt.title(f"RK45 - Evolución en {años} años")
    plt.xlabel("Tiempo (años)")
    plt.ylabel(r"$\Omega$ (rad/s)")
    plt.grid()
    plt.legend()
    plt.show()
```
