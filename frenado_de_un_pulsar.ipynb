### **¿Qué es un púlsar?**

Los pulsares son **estrellas de neutrones** altamente magnetizadas, que giran rápidamente sobre su propio eje. Aquellos que son jóvenes, rotan considerablemente más rápido que los longevos, puesto que pierden energía con el paso del tiempo. Esta pérdida de energía es objeto de estudio y es lo que analizaremos en el presente proyecto.

<div style="background-color: #f1f79d; color: black; padding: 10px; border-radius: 5px;">

### **Púlsar del Cangrejo**

El **Púlsar del Cangrejo (PSR B0531+21)** es uno de los pulsares mejor caracterizados observacionalmente. Está situado en la Nebulosa del Cangrejo, posee 25 km de diámetro y realiza 30 vueltas por segundo, o sea, su frecuencia es de 30 Hz. Su periodo de rotación disminuye 38 nanosegundos por día debido a la energía que libera. 

**¿Por qué ocurre esto?**

Si bien la rotación de este púlsar es muy estable debido a que no se aplica ninguna fuerza y su geometría no cambia, este frena con el paso del tiempo debido a la pérdida de energía. El principal mecanismo de pérdida de energía que opera sobre los pulsares es el **freno magnético**.


<div style="background-color: #f1f79d; color: black; padding: 10px; border-radius: 5px;">

### **Frenado magnetodipolar**

El modelo clásico de un púlsar describe a la estrella de neutrones como un dipolo magnético cuyo eje no coincide con el eje de rotación. Como consecuencia de esta desalineación, el momento dipolar varía en el tiempo, lo que conduce a la emisión de **radiación electromagnética**. La **potencia radiada**, equivalente a la pérdida de energía rotacional por unidad de tiempo, puede derivarse a partir de la electrodinámica clásica y depende del período de rotación y de la tasa de frenado del púlsar, ambas cantidades accesibles observacionalmente.

<div style="background-color: #d4eeff; color: black; padding: 10px; border-radius: 5px;">

### **El modelo físico**

La energía rotacional del púlsar es: 

$$

E_{rot} = \frac{1}{2}I(\Omega)(\Omega)^{2}

$$

donde 
- $I$ es el momento de inercia.
- $\Omega$ es la frecuencia angular de rotación.

Su evolución temporal se obtiene desde el balance energético: 

$$

\frac{dE_{rot}}{dt} = -P_{loss}

$$

donde $P_{loss}$ es la potencia total perdida por los procesos electrodinámicos.

### **El proyecto**

El momento de inercia se expresa en función de la frecuencia angular: 

$$

I(\Omega) = I_0(1 + \epsilon\Omega^2)

$$

La evolución del ángulo magnético es: 

$$

\alpha(t) = \alpha_0 + \beta t

$$

La potencia radiada por un dipolo magnético en rotación es: 

$$

P_{dip} = - \frac{2}{3c^{3}} (BR^{3})^{2}\Omega^{4}\sin^{2}{\alpha (t)}

$$

Por lo tanto, el **balance energético** es: 

$$

\frac{d}{dt} \left[ \frac{1}{2} I(\Omega) \Omega^2 \right] = -\frac{2B^2R^6}{3c^3} \Omega^4 \sin^2 \alpha(t)

$$

El **objetivo** del proyecto es resolver estas ecuaciones mediante los métodos numéricos RK4 y RK45, para posteriormente comparar los resultados arrojados con los valores observacionales, tomando en cuenta la estabilidad, precisión y el valor numérico del índice de frenado.

### **Preparar la ecuación diferencial**

Para poder resolver este problema numéricamente, primero debemos dejar la ecuación diferencial expresada de la siguiente forma:
$$
\frac{d \Omega(t)}{dt} = f(t,y)
$$
Sabemos que
$$
P_{dip} =  \frac{2}{3c^{3}} (BR^{3})^{2}\Omega^{4}\sin^{2}{\alpha (t)}
$$
Reemplazando en la ecuación
$$
\frac{d}{dt}\left[ \frac{1}{2}I(\Omega)\Omega^{2} \right] = -P_{dip}(t)
$$
Por definición $\large I(\Omega) = I_{0}(1+\epsilon\Omega^{2})$, reemplazando:
$$
\frac{d}{dt}\left[ \frac{1}{2}I_{0}(1+\epsilon\Omega^{2})\Omega^{2} \right] = -P_{dip}(t)
$$
Sacamos las constantes y distribuimos $\large \Omega^{2}$
$$
\begin{align}
\frac{d}{dt}[(1+\epsilon\Omega^{2})\Omega^{2}] &= -\frac{2P_{dip}}{I_{0}} \\
\frac{d}{dt}[(\Omega^{2}+\epsilon\Omega^{4})] &= -\frac{2P_{dip}}{I_{0}}
\end{align}
$$
Aplicamos la derivada usando la regla de la cadena
$$
\begin{align}
2\Omega\dot{\Omega} + 4\epsilon\Omega^{3}\dot{\Omega} &= -\frac{2P_{dip}}{I_{0}} \\
\dot{\Omega}(\Omega + 2\epsilon\Omega^{3}) &= -\frac{P_{dip}}{I_{0}}
\end{align}
$$
Finalmente obtenemos que:
$$
\boxed{\dot{\Omega} = \frac{d\Omega(t)}{dt} = -\frac{P_{dip}(t)}{I_{0}\Omega(1 + 2\epsilon\Omega^{2})}}
$$

La desigualdad $0 \le \sin^2\alpha(t) \le 1$ se cumple porque el seno de cualquier ángulo real siempre toma valores entre $-1$ y $1$. Al elevar ese valor al cuadrado, los números negativos se vuelven positivos y el máximo posible sigue siendo $1$, por lo que el resultado queda necesariamente entre $0$ y $1$. Por eso, $\sin^2\alpha(t)$ nunca puede ser negativo ni mayor que uno.

Esto nos sirve como condición para determinar constantes, si empezamos en $t =0 \implies \alpha(t=0) = \alpha_{0}$ por lo tanto:
$$
\implies \sin^{2}\alpha_{0} = -\frac{{\dot{\Omega_{0}}}{I_{0}(1+\epsilon 2\Omega_{0}^{2})}}{K\Omega_{0}^{3}}
$$
y este cómo debe estar entre cero y uno sólo nos sirven valores específicos de $\epsilon$ en este caso ocuparemos $\epsilon \approx 10^{-6}$. Si ocupamos los valores que nos dieron como iniciales nos da que:
$$
\alpha_{0} = \arcsin(\sqrt{ \sin^{2} \alpha_{0}}) \approx 0,64 \ \  [rad] 
$$
Finalmente el único valor que tabularemos es $\beta$ el cual para los pulsares es típico que estén en el orden de $10^{-13}$ y $10^{-12}$.

### 3. 

Definimos la función de la derivada respecto al tiempo de omega, para esto contruimos una función con todas las constantes necesarias.

```python
def omega_dot(t, y, B, R, c, I0, epsilon, alpha0, betta):
    # 'y' es una lista/array, extraemos el primer valor (omega)
    omega = y

    #Ángulo que depende del tiempo
    a = alpha0 + betta * t
    k = (2 * (B * R**3)**2) / (3 * c**3)

    #P_dip
    p_dip = k * (omega**4) * (np.sin(a)**2)
    denominador = I0 * omega * (1 + 2 * epsilon * (omega**2))
    
    f = -p_dip / denominador

    return f
```

#### 4.1 Primero partimos con la implementación de rk4 mediante nuestro modulo

```python
import module as mod

segundos_en_un_año = 365.25 * 24 * 3600 # Esto nos ayudará a expresar en años el eje temporañ
años_simulacion = [100, 1000, 10000] # Esto es una lista con los tiempos finales de cada resolución de la EDO

resultados_rk4 = {}

# Crear figura con subplots horizontales (ejes independientes)
fig, axes = plt.subplots(1, len(años_simulacion), figsize=(18,5))

for i, años in enumerate(años_simulacion): #Acá haremos una resolución variando los años 
    t_final = años * segundos_en_un_año
    h_estable = 24 * 3600

    solucion_t, omega_rk4 = mod.rk4_method(
        lambda t, y: omega_dot(t, y, B, R, c, I0, epsilon, alpha0, betta),
        t0=0.0,
        y0=omega0,
        h=h_estable,
        tf=t_final
    )

    resultados_rk4[años] = (solucion_t, omega_rk4)
    t_años = solucion_t / segundos_en_un_año

    # Graficar en el subplot correspondiente
    ax = axes[i]
    ax.plot(t_años, omega_rk4, color="royalblue", label="RK4")
    ax.set_title(f"{años} años")
    ax.set_xlabel("Tiempo (años)")
    ax.set_ylabel(r"$\Omega$ (rad/s)")
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.legend()

plt.tight_layout()
plt.show()
```

#### 4.2 Ahora implementaremos la solución de scipy con el método rk45

```python
from scipy.integrate import solve_ivp

# Preparamos la figura y el diccionario
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
resultados_rk45 = {}

for i, años in enumerate(años_simulacion):
    # Usamos el tiempo de la celda anterior
    solucion_t, _ = resultados_rk4[años]
    t_final = años * segundos_en_un_año

    # Aquí pasamos omega_dot directamente porque Scipy ya maneja listas
    omega_rk45 = solve_ivp(
        fun=omega_dot,
        t_span=(0, t_final),
        y0=[omega0],
        method="RK45",
        args=(B, R, c, I0, epsilon, alpha0, betta),
        rtol=1e-9,
        atol=1e-12,
        t_eval=solucion_t
    )

    # Guardamos los resultados por si los necesitas después
    resultados_rk45[años] = omega_rk45
    t_años = omega_rk45.t / segundos_en_un_año

    # Graficamos
    ax = axes[i]
    ax.plot(t_años, omega_rk45.y[0], label="RK45 (Scipy)", color="darkorange", linewidth=2)
    
    ax.set_title(f"RK45 Scipy - {años} años")
    ax.set_xlabel("Tiempo (años)")
    ax.set_ylabel(r"$\Omega$ (rad/s)")
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.legend()

plt.tight_layout()
plt.show()
```

El índice de frenado (n) es una cantidad adimensional que describe cómo depende la desaceleración de la velocidad angular.
 Se define como 
$$ 
n = \frac{d\ln|\dot{\Omega}|}{d\ln\Omega}

$$
donde, como hemos visto, $\Omega$ es la velocidad angular y $\dot{\Omega}$ es la aceleración angular.

Procederemos a calcular $n$.

### **Error relativo**

El error relativo se utiliza para cuantificar la diferencia entre las soluciones obtenidas mediante los métodos numéricos RK4 y RK45. Se define como:

$$
\text{error}_{\text{rel}} =
\frac{\left|\Omega_{\text{RK4}} - \Omega_{\text{RK45}}\right|}
{\left|\Omega_{\text{RK45}}\right|}.
$$

Esta cantidad permite evaluar qué tan cercanos son los resultados de ambos métodos en cada instante de tiempo. Valores pequeños del error relativo indican que las dos integraciones producen soluciones prácticamente indistinguibles, lo que sugiere que la solución numérica es estable y que el paso temporal utilizado es adecuado para describir la evolución rotacional del púlsar.


```python
# =========================
# error relativo
# =========================
seguro = 1e-300  # para evitar divisiones por cero

omega_rk45_vals = omega_rk45.y[0]

error_rel = np.abs(omega_rk4 - omega_rk45_vals) / (np.abs(omega_rk45_vals) + seguro)

print("Error relativo máximo =", np.max(error_rel))
print("Error relativo promedio =", np.mean(error_rel))

# =========================
# recorte de bordes
# =========================
k = 20
t_mid_años = t_años[k:-k]
error_mid = error_rel[k:-k]

# =========================
# gráfico del error relativo
# =========================
plt.figure()
plt.plot(t_mid_años, error_mid)
plt.xlabel("Tiempo (años)")
plt.ylabel("Error relativo")
plt.title("Error relativo entre RK4 y RK45")
plt.show()
```

## Análisis de Hipótesis para la Evolución del Púlsar

A continuación plantearemos variaciones en los supuestos físicos para contrastar con el modelo base del Púlsar del Cangrejo. (Para las hipótesis se usará el método de rk45 de acipy por comodidad)

#### 1. Hipótesis de Estrella de Neutrones Rígida ($\epsilon = 0$ y $\alpha(t) = \frac{\pi}{2}$)

En esta configuración, se asume que la estrella es un cuerpo rígido perfecto, ignorando cualquier deformación estructural provocada por la rotación. Bajo esta premisa, el momento de inercia se mantiene constante $I=I_0​$, lo que simplifica drásticamente la ecuación de balance energético al eliminar los términos de deformación centrífuga.

Para esta comprobación, fijamos el ángulo de inclinación en α=π/2. Esto representa un púlsar ortogonal, donde el eje magnético es perfectamente perpendicular al eje de rotación. Físicamente, esto implica que la potencia de frenado por radiación dipolar es máxima, ya que $sin^{2}(π/2)=1$ esto teóricamente debería dar como resultado $n=3$.

#### 2. Hipótesis de Decaimiento del Campo Magnético ($B = B(t)$)

En el modelo presentado se asume un campo magnético fijjo: $3.8×1012 \ [G]$ esto no es lo común en sistemas físicos complejos queremos proponer que este campo también evoluciona. es por esto que podemos usar, por ejemplo, un decaimiento exponencial como $B(t)=B_{0}​e^{−t/\tau}$ para simular cómo el púlsar va "envejeciendo" y perdiendo fuerza magnética en su superficie.

Esto implica que, al tener un campo cada vez más débil, la eficiencia del torque de frenado disminuye a medida que pasa el tiempo. Normalmente, esto resulta en un índice de frenado $n>3$.

## Hipótesis 1

Partimos del modelo para el momento de inercia dependiente de la velocidad angular
$$
I(\Omega) = I_0(1+\epsilon\Omega^2)
$$

Cuando $\epsilon = 0$ entonces el momento de inercia es constante
$$
I(\Omega) = I_0
$$
y la ecuación de evolución de la velocidad angular se reduce a
$$
\dot{\Omega} = -\frac{P_{dip}}{I_0\Omega}
$$
---


Cómo sabemos la potencia radiada por un dipolo magnético rotante está dada por
$$
P_{dip} = \frac{2}{3c^3}(BR^3)^2\Omega^4\sin^2\alpha
$$

donde $B$ es el campo magnético superficial, $R$ es el radio de la estrella y $c$ es la velocidad de la luz.

Si el eje magnético es perpendicular al eje de rotación es decir $\alpha (t) = \pi / 2$ entonces:
$$
\sin^2\alpha = 1
$$
y la potencia radiada se simplifica a
$$
P_{dip} = \frac{2}{3c^3}(BR^3)^2\Omega^4
$$
Sustituyendo en la ecuación de evolución obtenemos
$$
\dot{\Omega}
=
-\frac{1}{I_0\Omega}
\left(
\frac{2}{3c^3}(BR^3)^2\Omega^4
\right)
$$
lo que simplifica a
$$
\dot{\Omega}
=
-\frac{2(BR^3)^2}{3c^3 I_0}\Omega^3
$$
Definiendo la constante
$$
K = \frac{2(BR^3)^2}{3c^3 I_0}
$$
la ecuación final queda
$$
\dot{\Omega} = -K\Omega^3
$$
---

### Índice de frenado

El índice de frenado además se puede definir como
$$
n = \frac{\Omega\ddot{\Omega}}{\dot{\Omega}^2}
$$

El cual es equivalente con la expresión anterior del indice de frenado. Para el caso:
$$
\dot{\Omega} = -K\Omega^3
$$
se obtiene directamente
$$
n = 3
$$
Por lo tanto, cuando $\epsilon = 0$ y $\alpha = \pi/2$, el frenado está completamente dominado por radiación dipolar magnética y el valor teórico esperado del **braking index** es
$$
n = 3
$$

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
resultados_epsilon_0 = {}

for i, años in enumerate(años_simulacion):
    # Usamos el tiempo de la celda anterior
    solucion_t, _ = resultados_rk4[años]
    t_final = años * segundos_en_un_año

    # Aquí pasamos omega_dot directamente porque Scipy ya maneja listas
    omega_epsilon_0 = solve_ivp(
        fun=omega_dot,
        t_span=(0, t_final),
        y0=[omega0],
        method="RK45",
        args=(B, R, c, I0, epsilon_0, alpha_pi2, betta_0),
        rtol=1e-9,
        atol=1e-12,
        t_eval=solucion_t
    )

    # Guardamos los resultados por si los necesitas después
    resultados_epsilon_0[años] = omega_epsilon_0
    t_años = omega_epsilon_0.t / segundos_en_un_año

    # Graficamos
    ax = axes[i]
    ax.plot(t_años, omega_epsilon_0.y[0], label="$\epsilon = 0$", color="black", linewidth=2)
    
    ax.set_title(f"RK45 Scipy - {años} años")
    ax.set_xlabel("Tiempo (años)")
    ax.set_ylabel(r"$\Omega$ (rad/s)")
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
# =========================
# Cálculo del braking index para epsilon = 0
# =========================

seguro = 1e-300

fig, axes = plt.subplots(1, 3, figsize=(18,5))

for i, años in enumerate(años_simulacion):

    sol = resultados_epsilon_0[años]

    t = sol.t
    omega = sol.y[0]

    # calcular omega_dot numéricamente
    omega_dot = np.gradient(omega, t)

    # logs
    lnOmega = np.log(omega + seguro)
    lnAbsDomega = np.log(np.abs(omega_dot) + seguro)

    # derivadas
    d_lnOmega_dt = np.gradient(lnOmega, t)
    d_lnAbsDomega_dt = np.gradient(lnAbsDomega, t)

    # braking index
    n = d_lnAbsDomega_dt / (d_lnOmega_dt + seguro)

    # recorte de bordes
    k = 20
    t_mid = t[k:-k] / segundos_en_un_año
    n_mid = n[k:-k]

    # gráfico
    ax = axes[i]
    ax.plot(t_mid, n_mid, color="black", label=r"$\epsilon=0,\ \alpha=\pi/2$")
    ax.axhline(3, linestyle=":", color="gray", label="n = 3 (teórico)")

    ax.set_title(f"Braking index - {años} años")
    ax.set_xlabel("Tiempo (años)")
    ax.set_ylabel("n")
    ax.grid(True, linestyle="--", alpha=0.7)
    ax.legend()

plt.tight_layout()
plt.show()
```

# Hipótesis 2: Decaimiento del Campo Magnético ($B = B(t)$)

En el modelo de frenado estándar se asume un campo magnético superficial constante (e.g., $B_0 = 3.8 \times 10^{12}\text{ G}$). Sin embargo, en sistemas astrofísicos complejos, es razonable proponer que este campo no es estático y evoluciona con el púlsar. Para modelar el "envejecimiento" de la estrella de neutrones y la progresiva pérdida de intensidad magnética en su superficie, introducimos un decaimiento de tipo exponencial:

$$B(t) = B_{0}e^{-t/\tau}$$

Para acoplar esta dinámica al modelo, agrupamos las constantes estructurales definiendo $C = \frac{2R^{6}}{3c^{3}}$. Reemplazando esta constante en la ecuación de balance energético, la derivada de la frecuencia angular toma la siguiente forma general:

$$\dot{\Omega} = -\frac{C\Omega^{3}}{I_{0}(1 + 2\epsilon\Omega^{2})}B^{2}(t)\sin^{2}\alpha(t)$$


### Constante Temporal

* **La constante temporal $\tau$:** Representa el tiempo característico de decaimiento del campo magnético; es decir, el tiempo necesario para que la intensidad del campo se reduzca a un $1/e$ (aproximadamente el **37%**) de su valor original $B_0$. En la implementación numérica computacional, configurar un valor como $\tau = 10^{10}$ permite simular un decaimiento más lento, asegurando la estabilidad del modelo a largo plazo.

<div style="background-color: #d4eeff; color: black; padding: 10px; border-radius: 5px;">

Para integrar la evolución temporal de $B$, se reestructuró la función "omega_dot()" sustituyendo el parámetro estático por la constante de tiempo $\tau$. La implementación se realizó de forma modular mediante una función "B_variable" que toma tanto al tiemo como a tau dentro de la función "omega_dot()"

```python
def B_variable(t,tau): # definimos nuestro tau
    B0 = 3.8e12
    return B0 * np.exp(-t / tau) 

def omega_dot_B(t, y, R, c, I0, epsilon, alpha0, betta, tau):
    # 'y' es una lista/array, extraemos el primer valor (omega)
    omega = y

#- - - - -  - - agregmos el campo magnetico variable- - - - - - -- - - - - -- - - - -
    B = B_variable(t,tau)


    #Ángulo que depende del tiempo
    a = alpha0 + betta * t
    k = (2 * (B * R**3)**2) / (3 * c**3)

    #P_dip
    p_dip = k * (omega**4) * (np.sin(a)**2)
    denominador = I0 * omega * (1 + 2 * epsilon * (omega**2))
    
    f = -p_dip / denominador

    return f


# Constante tau
tau = 1e10
```

<div style="background-color: #d4eeff; color: black; padding: 10px; border-radius: 5px;">


Ahora calularemos este omega con el campo magnético variable

```python
# Preparamos la figura y el diccionario
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
resultados_rk45 = {}

for i, años in enumerate(años_simulacion):
    # Usamos el tiempo de la celda anterior
    solucion_t, _ = resultados_rk4[años]
    t_final = años * segundos_en_un_año

    omega_sol = solve_ivp(
        fun=omega_dot_B,
        t_span=(0, t_final),
        y0=[omega0],
        method="RK45",
        args=(R, c, I0, epsilon, alpha0, betta, tau),
        t_eval=solucion_t)

    # Guardamos los resultados por si los necesitas después
    resultados_rk45[años] = omega_sol
    t_años = omega_sol.t / segundos_en_un_año

    # Graficamos
    ax = axes[i]
    ax.plot(t_años, omega_sol.y[0], label="$\Omega$ con $B(t)$", color="red", linewidth=2)
    
    ax.set_title(f"$\Omega$ con $B(t)$ - {años} años")
    ax.set_xlabel("Tiempo (años)")
    ax.set_ylabel(r"$\Omega$ (rad/s)")
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
# =========================
# Cálculo del braking index para B(t)
# =========================

seguro = 1e-300

fig, axes = plt.subplots(1, 3, figsize=(18,5))

for i, años in enumerate(años_simulacion):

    sol = resultados_rk45[años]

    t = sol.t
    omega = sol.y[0]

    # calcular omega_dot numéricamente
    omega_dot = np.gradient(omega, t)

    # logs
    lnOmega = np.log(omega + seguro)
    lnAbsDomega = np.log(np.abs(omega_dot) + seguro)

    # derivadas
    d_lnOmega_dt = np.gradient(lnOmega, t)
    d_lnAbsDomega_dt = np.gradient(lnAbsDomega, t)

    # braking index
    n = d_lnAbsDomega_dt / (d_lnOmega_dt + seguro)

    # recorte de bordes
    k = 20
    t_mid = t[k:-k] / segundos_en_un_año
    n_mid = n[k:-k]

    # promedio
    n_prom = np.mean(n_mid)

    print(f"{años} años -> n promedio = {n_prom:.5f}")

    # gráfico
    ax = axes[i]

    ax.plot(t_mid, n_mid, color="red", label=r"$B(t)$")

    ax.set_title(f"Braking index con $B(t)$ - {años} años")
    ax.set_xlabel("Tiempo (años)")
    ax.set_ylabel("n")

    ax.grid(True, linestyle="--", alpha=0.7)
    ax.legend()

plt.tight_layout()
plt.show()
```

Las simulaciones realizadas para un campo magnético con decaimiento exponencial ($B(t) = B_0 e^{-t/\tau}$) muestran un comportamiento divergente en el índice de frenado ($n$), permitiendo extraer las siguientes conclusiones:

#### 1. Incremento de $n$ sobre el Límite Teórico
En la escala de **100 años**, el índice de frenado aumenta de forma sostenida desde $n \approx 19$ hasta $n \approx 34$. Este fenómeno ocurre porque el debilitamiento del campo magnético reduce la eficiencia del torque de frenado. Al disminuir la desaceleración angular ($|\dot{\Omega}|$), la curvatura de la evolución de $\Omega$ se aplana, lo que matemáticamente se traduce en un índice $n > 3$ como se puede comprobar $n \approx 25,8$

#### 2. Inconsistencia con la Observación Real
El Púlsar del Cangrejo presenta un índice de frenado observado de $n_{obs} \approx 2.51$ los resultados de esta hipótesis ($n \gg 3$) permiten concluir que el decaimiento del campo magnético no es el mecanismo responsable del frenado en púlsares jóvenes puesto que se aleja demaciado a lo modelado anteriormente.

#### 3. Colapso del Modelo y Límite Numérico
En la escala de 1,000 años el índice $n$ escala a valores del orden de $40,000$. Físicamente, esto indica que el campo ha decaído tanto que la estrella casi no experimenta frenado, haciendo que el cálculo de $n$ (que depende del inverso de $\dot{\Omega}$) diverja.

En la escala de 10,000 años se observa ruido numérico extremo ($10^{297}$). Esto representa el límite de precisión de punto flotante, el torque es tan despreciable que las derivadas se vuelven estocásticas, marcando el fin de la validez del modelo.

#### Conclusión de la Hipótesis
La variabilidad del campo magnético (con el $\tau$ seleccionado) genera una evolución incompatible con el pulsar del cangrejo. Para futuros ajustes, se requeriría un $\tau$ significativamente mayor o considerar que el campo magnético permanece casi estático durante la fase de vida temprana del púlsar.
